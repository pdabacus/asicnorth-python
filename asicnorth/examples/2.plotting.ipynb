## 2D Arrays
---

```python
print("5 x 10 matrices\n")

my_2d_list = [[10*i+j for j in range(10)] for i in range(5)]

print("2d python list")
for i in range(len(my_2d_list)):
    print(f"row{i} = {my_2d_list[i]}")
    
print("\nnumpy matrix")
my_2d_arr = np.matrix(my_2d_list)
print(my_2d_arr)
```

---
<a id="saveload"></a>
## Saving and Loading
---

```python
# save with native python

f = open("data.txt", "w")
for row in my_2d_list:
    row_str_arr = ["%2d"%num for num in row]
    row_str = (",").join(row_str_arr)
    print(row_str, file=f)
f.close()

!cat data.txt
```

```python
# load with native python

loaded_2d_list = list()

f = open("data.txt", "r")
line = f.readline()
while line != '':
    row_str_arr = line.strip().split(",")
    row = [int(num_str) for num_str in row_str_arr]
    loaded_2d_list.append(row)
    line = f.readline()
f.close()

print("parsed list from file")
for i in range(len(loaded_2d_list)):
    print(f"row{i} = {loaded_2d_list[i]}")
         
```

```python
# save using numpy
np.savetxt("data.txt", my_2d_arr, fmt="%2d", delimiter=",")

!cat data.txt
```

```python
# load using numpy
loaded_2d_arr = np.loadtxt("data.txt", delimiter=",", dtype="int")

print(loaded_2d_arr)
```

## Optimization
---

```python
def wiggly(x):
    return np.square(x) + 1 - np.cos(10*x)

x = np.arange(-5, 5, 0.01)
y = wiggly(x)
plt.plot(x, y)
plt.title("wiggly(x)")
plt.show()

print(f"wiggly(0.0) = {wiggly(0.0)}")
```

```python
guess_min = 5.0
minimum = scipy.optimize.minimize(wiggly, guess_min)
print(minimum)
```

```python
from pavan_optimize import GradientDescent

# plot function and current point in gradient descent
def plot_pt(x):
    X = np.arange(-5, 5, 0.01)
    Y = wiggly(X)
    plt.plot(X, Y)
    plt.plot([x], [wiggly(x)], "rx")
    plt.show()

plot_pt(guess_min)
g = GradientDescent(wiggly, guess_min, mass=0.7)

for i in range(5):
    g.descend(5, alpha=lambda t:0.01+0.02/t)
    x = g.parameters()[0]
    print("Iteration %2d: x=%s" % (i+1, x))
    plot_pt(x)

```

```python
from pavan_optimize import GeneticMinimization

# plot function and population of points in genetic algorithm
def plot_pop(pop, score):
    X = np.arange(-5, 5, 0.01)
    Y = wiggly(X)
    plt.plot(X, Y)
    plt.plot(pop, score, "rx")
    plt.show()

g = GeneticMinimization(wiggly, (-4,4), pop_size=40, trn_size=10)

for i in range(5):
    g.breed()
    x, err  = g.best()
    print("Iteration %2d: x=%s" % (i+1, x[0]))
    plot_pop(g.population(i), g.population_scores(i))

```
